```python
import pandas as pd
from statsbombpy import sb
import warnings

warnings.filterwarnings('ignore')

print("✅ Librerie caricate. Riprovo lo scaricamento...")

# 1. lista delle partite (Mondiale 2018)
matches = sb.matches(competition_id=43, season_id=3)
all_shots = []

print(f"Trovate {len(matches)} partite. Analizzo le prime 5 per fare veloce...")

# prendiamo le prime 5 partite
for match_id in matches['match_id'][:5]:
    try:
        # Scarica eventi
        events = sb.events(match_id=match_id)
        
        match_shots = events[events['type'] == 'Shot'].copy()
        
        if not match_shots.empty:
            all_shots.append(match_shots)
            print(f"   -> Partita {match_id}: Trovati {len(match_shots)} tiri.")
            
    except Exception as e:
        print(f"   -> Errore nella partita {match_id}: {e}")

if all_shots:
    df = pd.concat(all_shots, ignore_index=True)
    
    print("\n" + "="*50)
    print(" REPORT DI FATTIBILITÀ (DA USARE PER LA MAIL) ")
    print("="*50)
    print(f"Tiri totali nel campione: {len(df)}")
    
    # --- VERIFICA  ---
    print("\n[VERIFICA REQUISITI CLUB vs NAZIONALI]")
    
    # 1. Play Pattern (Azione manovrata vs Piazzato)
    if 'play_pattern' in df.columns:
        print(f"✅ Variabile 'play_pattern' -> PRESENTE")
        print(f"   (Valori trovati: {df['play_pattern'].unique()})")
    else:
        print(f"❌ Variabile 'play_pattern' -> ASSENTE")

    # 2. Under Pressure (Pressione difensiva)
    if 'under_pressure' in df.columns:
        print(f"✅ Variabile 'under_pressure' -> PRESENTE")
        # Conta quanti tiri sono sotto pressione
        pressati = df['under_pressure'].sum() if df['under_pressure'].dtype != 'O' else df['under_pressure'].count()
        print(f"   (Trovati tiri sotto pressione nel campione)")
    else:
        print(f"⚠️ Variabile 'under_pressure' -> ASSENTE in questo campione (verificare su più dati)")
        
    # 3. Shot Type (Per togliere i rigori)
    if 'shot_type' in df.columns:
        print(f"✅ Variabile 'shot_type' -> PRESENTE")
    
else:
    print("⚠️ Ancora nessun dato. C'è qualcosa di strano nella connessione o nei dati.")
```
